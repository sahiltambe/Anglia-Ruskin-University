```python
import math
import matplotlib.pyplot as plt
import networkx as nx

# Step 1: Define Student ID and Compute Inputs
# The values are calculated using the given formula and SID
SID = 2353465
denominator = 8964879

x1 = (SID / denominator) * 23
x2 = (SID / denominator) * 32
x3 = (SID / denominator) * 56
x4 = (SID / denominator) * 48

# Step 2: Define Weights for Hidden Layer 1
# Each input is connected to H1 and H2 with a weight of 0.2
weights_H1_H2 = [0.2] * 8  # w1 to w8

# Compute Activation for Hidden Layer 1 (H1, H2)
z_H1 = sum(weights_H1_H2[i] * [x1, x2, x3, x4][i % 4] for i in range(0, 8, 2))
z_H2 = sum(weights_H1_H2[i] * [x1, x2, x3, x4][i % 4] for i in range(1, 8, 2))

# ReLU Activation Function
def relu(z):
    return max(0, z)

H1, H2 = relu(z_H1), relu(z_H2)

# Step 3: Define Weights for Hidden Layer 2
weights_H3_H4 = [0.1] * 4  # w9 to w12

# Compute Activation for Hidden Layer 2 (H3, H4)
z_H3 = (weights_H3_H4[0] * H1) + (weights_H3_H4[2] * H2)
z_H4 = (weights_H3_H4[1] * H1) + (weights_H3_H4[3] * H2)

H3, H4 = relu(z_H3), relu(z_H4)

# Step 4: Define Weights for Output Layer
weights_O1_O2 = [0.3] * 4  # w13 to w16
bias_weight = 0.5

# Compute Activation for Output Layer (O1, O2)
z_O1 = (weights_O1_O2[0] * H3) + (weights_O1_O2[2] * H4) + bias_weight
z_O2 = (weights_O1_O2[1] * H3) + (weights_O1_O2[3] * H4) + bias_weight

# Sigmoid Activation Function
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

O1, O2 = sigmoid(z_O1), sigmoid(z_O2)

# Display Final Outputs
print("Final Output O1:", O1)
print("Final Output O2:", O2)

# Step 5: Plot Activation Functions
z_values = [i for i in range(-10, 11)]
sigmoid_values = [sigmoid(z) for z in z_values]
relu_values = [relu(z) for z in z_values]

plt.figure(figsize=(10, 5))
plt.plot(z_values, sigmoid_values, label="Sigmoid", color='blue')
plt.plot(z_values, relu_values, label="ReLU", color='red')
plt.xlabel("Input (z)")
plt.ylabel("Activation Output")
plt.title("Activation Function Comparison")
plt.legend()
plt.grid()
plt.show()
```
